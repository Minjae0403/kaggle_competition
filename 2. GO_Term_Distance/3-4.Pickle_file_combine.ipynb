# File Download function

```python
import os
import subprocess
from IPython.display import FileLink, display

def download_file(path, download_file_name):
    os.chdir('/kaggle/working/')
    zip_name = f"/kaggle/working/{download_file_name}.zip"
    command = f"zip {zip_name} {path} -r"
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    if result.returncode != 0:
        print("Unable to run zip command!")
        print(result.stderr)
        return
    display(FileLink(f'{download_file_name}.zip'))
    

# download_file('/kaggle/working/distance_tensor_1to2.pkl', 'distance_tensor_1to2')  
```

# Pickle File Load

```python
import pickle

file_path_1 = '/kaggle/input/distance-tensor-1/distance_tensor_1.pkl'
with open(file_path_1, 'rb') as file:
    tensor_1 = pickle.load(file)


file_path_2 = '/kaggle/input/distance-tensor-2/distance_tensor_2.pkl'
with open(file_path_2, 'rb') as file:
    tensor_2 = pickle.load(file)


print(tensor_1.shape, tensor_2.shape)
```

# Tensor Concanate and Pickle Save

```python
import torch
distance_tensor_1to2 = torch.cat((tensor_1, tensor_2), dim=0)
print(distance_tensor_1to2.shape)
```

```python
with open('distance_tensor_1to2.pkl', 'wb') as f:
	pickle.dump(distance_tensor_1to2, f)
```
